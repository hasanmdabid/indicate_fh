```python
# List of image and filter names
images = ['original_0', 'original_1', 'original_2', 'original_3', 'original_4']
filters = ['xception', 'mobilenetv2', 'vgg16', 'inceptionv3']

# Create a figure for the matrix of subplots
fig, axes = plt.subplots(len(images), len(filters) + 1, figsize=(20, 20))

# Plot original images in the first column
for image_index, image_name in enumerate(images):
    file_path = os.path.join(filtered_images_directory, f"{image_name}.png")

    # Load the image
    img = load_img(file_path)
    img_array = img_to_array(img)

    # Display the image in the appropriate subplot
    axes[image_index, 0].imshow(img_array.astype('uint8'))
    axes[image_index, 0].axis('off')
    if image_index == 0:
        axes[image_index, 0].set_title("Original", fontsize=12)

# Plot filtered images
for filter_index, filter_name in enumerate(filters):
    for image_index in range(len(images)):
        # Construct the filename for the filtered image
        filename = f"{filter_name}_{image_index}.png"
        file_path = os.path.join(filtered_images_directory, filename)

        # Load the image
        img = load_img(file_path)
        img_array = img_to_array(img)

        # Display the image in the appropriate subplot
        axes[image_index, filter_index + 1].imshow(img_array.astype('uint8'))
        axes[image_index, filter_index + 1].axis('off')

        # Set the filter name as the column title
        if image_index == 0:
            axes[image_index, filter_index +
                 1].set_title(filter_name, fontsize=12)

plt.tight_layout()
plt.show()
```
